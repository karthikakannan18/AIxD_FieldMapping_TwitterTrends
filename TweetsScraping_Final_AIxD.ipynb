```python
#import the libraries
import twint
import time
import pandas as pd
import nest_asyncio
nest_asyncio.apply()
```

```python
#keywords are username of twitter accounts- AI and tech leaders(AI and Design)
twitter_keywords = ['cynthiabreazeal', 'GoogleAI','AndrewYNg', 'OpenAI', 'jhamrick', 'TeachTheMachine', 'rasbt', 'lishali88', 'sharma_kriti', 'mike_matas', 'hardmaru', 'Soulegit', 'jenniferdaniel','PatrickHebron', 'dribnet', 'samim']
```

```python
# looping through each account using same configuration between the year 2017-01-01 to 2022-09-28
for username in twitter_keywords:
 c = twint.Config()
 c.Lang = "en"
 c.Username = username
 c.Since = '2017-01-01'
 c.Until = '2022-09-28'
 c.Pandas = True
 c.Store_pandas = True
 c.Pandas_au = True
 c.Pandas_clean = False
 twint.run.Search(c)
# store in data frame
tweets_df= twint.storage.panda.Tweets_df
```

```python
#saving to csv
tweets_df.to_csv('tweetsbetween2017-22.csv', index = False, header=True)
```

```python
#finding the size of data
tweets_df.shape
```
